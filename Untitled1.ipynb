```python
# @title Colab Setup and Imports { display-mode: "form" }
# @markdown (double click to see the code)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
# @markdown 默认损失函数
use_local_fuction = True # @param {type:"boolean"}
user_cost_function = "1 / (s * np.sin(s) + 12)"# @param {type:"string"}
# @markdown 优化变量取值范围
up_border = 12.55 # @param {type:"number"}
down_border = 0 # @param {type:"number"}


if use_local_fuction:
    cost_function = "1 / (s * np.sin(s) + 12+s)"
else:
    cost_function = user_cost_function
#定义损失函数

def Cost(s):
    return eval(cost_function)
```

show cost_function image

```python
xx = np.linspace(down_border,up_border,100)
yy = Cost(xx)
plt.figure(dpi=100,figsize=(10,6))
plt.plot(xx,yy)
plt.title("cost_function")
```

```python
t0 = np.var(np.random.uniform(0, 12.55, 100))


# @markdown 默认初始值
use_local_s0 = True # @param {type:"boolean"}
# @markdown 用户自定义初始值
user_s0 = 0 #@param {type:"number"}
if use_local_s0:
    s0 = np.random.uniform(down_border, up_border, 1)
else:
    s0 = 0
# @markdown 设定迭代次数
iters = 300 # @param {type:"number"}

# @markdown 设定终止条件，连续ct个新解都没有接受（ct/cn大于某个值）时终止算法
ct = 20 # @param {type:"number"}
```

# **SA 模拟退火算法迭代**

```python

%matplotlib inline
from IPython import display
# 保存历史最好的状态，默认取上边界值
best = up_border




# 启用交互模式
plt.ion()

def plot_durations(s0,keep=False):
    plt.figure(dpi=100,figsize=(10,6))
    plt.plot(xx,yy)
    plt.scatter(s0, Cost(s0),color='g')
    plt.title("cost_function")
    plt.pause(0.05)  # 暂停以更新图形
    if(keep):
        return
    else:
        display.clear_output(wait=True)

ct_array = []

# 保存历史最好的状态，默认取上边界值
best = 12.55
plot_durations(s0)
for t in range(1, iters + 1):
    # 在s0附近，产生新解，但又能包含定义内的所有值
    s1 = np.random.normal(s0, 2, 1)
    while s1 < 0 or s1 > 12.55:
        s1 = np.random.normal(s0, 2, 1)
    # 计算能量增量
    delta_t = Cost(s1) - Cost(s0)
    if delta_t < 0:
        s0 = s1
        ct_array.append(1)
    else:
        p = np.exp(-delta_t / t0)
        if np.random.uniform(0, 1) < p:
            s0 = s1
            ct_array.append(1)
        else:
            ct_array.append(0)

    plot_durations(s0)
    best = s0 if Cost(s0) < Cost(best) else best

    # 更新温度
    t0 = t0 / np.log(1 + t)

    # 检查终止条件
    if len(ct_array) > ct and np.sum(ct_array[-ct:]) == 0:
        print("迭代 ", t, " 次，连续 ", ct, " 次没有接受新解，算法终止！")
        break
plot_durations(best,True)
print("最优点：s=",best,"最小损失：Cost=",Cost(best))
```
